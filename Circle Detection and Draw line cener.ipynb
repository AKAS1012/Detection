```python
# create a function for visualization of given data 
def plot_images(img1, img2, title1="", title2=""):
    fig = plt.figure(figsize=(15,15))
    ax1 = fig.add_subplot(121)
    ax1.imshow(img1,cmap="gray")
    ax1.set(xticks =[],yticks=[], title=title1)

    fig = plt.figure(figsize=(15,15))
    ax2 = fig.add_subplot(121)
    ax2.imshow(img2,cmap="gray")
    ax2.set(xticks =[],yticks=[], title=title2)
```

```python
# Convert to grayscale.
output = img.copy()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Blur using 3 * 3 kernel. 
blurred = cv2.medianBlur(gray, 5)
# Apply Hough transform on the blurred image. 
detected_circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, 1, 100,  param1=240,minRadius=0, maxRadius=0)
# if param and Radius change then circle detected accurate
```

```python
#sure at least some circles were found
if detected_circles is not None:
	# convert the (x, y) coordinates and radius of the circles to integers
	 detected_circles = np.uint16(np.around(detected_circles))
    # loop over the (x, y) coordinates and radius of the circles
	 for (x, y, r) in detected_circles[0,:2]: # please slice the iloc
		# draw the circle in the output image, then draw a rectangle
		# corresponding to the center of the circle
		 cv2.circle(output, (x, y), r, (0, 255, 0), 2)
		 cv2.circle(output, (x , y), 2, (0, 0, 255), -1)
      
        


        
       
cv2.imshow("output", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('rupee.jpeg',0) # change image to gray scale
img = cv2.medianBlur(img, 5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR) # change color gray to BGR

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,
                         param1=63,minRadius=9)
if circles is not None:
    circles = np.round(circles[0,:6]).astype("int") 

    for (x, y, r) in circles:
        # Draw the circle in the output image
        cv2.circle(cimg, (x, y), r, (0,255,0), 3)
        # Draw a rectangle(center) in the output image
        cv2.rectangle(cimg, (x - 2, y - 2), (x + 2, y + 2), (0,255,0), -1)




cv2.imshow('detected circles',cimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
